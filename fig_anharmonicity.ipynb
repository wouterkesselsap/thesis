```python
from qutip import *
from supports import *
from plotsettings import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nq = 10         # number of levels in qubit
wq = 6.5 *2*pi  # bare g-e qubit transition frequency
Ec = 0.2 *2*pi  # anharmonicity (charging energy)
```

```python
b, nq= ops(Nq)
H_Kerr = wq*nq - Ec/2*b.dag()*b.dag()*b*b
H_Duff = (wq+Ec)*nq - Ec/12*(b + b.dag())**4

EE_Kerr = H_Kerr.eigenenergies()
EE_Duff = H_Duff.eigenenergies()

trans_Kerr = np.diff(EE_Kerr)
trans_Duff = np.diff(EE_Duff)
```

```python
plt.figure(figsize=(10,5))
plt.plot(np.asarray(range(len(trans_Duff)))+1, trans_Duff/2/pi, 'k--', label="Duffing")
plt.plot(np.asarray(range(len(trans_Kerr)))+1, trans_Kerr/2/pi, 'k', label="Kerr")
plt.xlim([1,9])
plt.xlabel('$i$')
plt.xticks(range(1,10))
plt.ylabel('$\\omega$ $/2\\pi$')
plt.title("$i^{{th}}$ transition frequency")
plt.legend()
plt.show()
```
